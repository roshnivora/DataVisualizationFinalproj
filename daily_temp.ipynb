```python
import pandas as pd
import numpy as np
import os as os
import altair as alt
```

I need to import the data. 

```python
PATH = '/Users/Carl/Documents/gitfiles/DataVisualizationFinalproj'
temp_data_path = os.path.join(PATH, 'USW00014819.csv')
temp_df = pd.read_csv(temp_data_path)
temp_df
```

```python
# Convert temp to datetime
temp_df['DATE'] = pd.to_datetime(temp_df['DATE'])

# Define function to convert celsius to fahrenheit
def c_to_f(df, col_name, new_col_name=None):
   
    if new_col_name is None:
        new_col_name = col_name + '_F'
        
    df[new_col_name] = ((df[col_name])/10) * 9/5 + 32
    return df

temp_df = c_to_f(temp_df, 'TAVG')

temp_df = c_to_f(temp_df, 'TMAX')

temp_df = c_to_f(temp_df, 'TMIN')

temp_df['max_min_avg'] = (temp_df['TMAX_F'] + temp_df['TMIN_F'])/2

temp_df
```

```python
# Define start and end dates
start_date = "2016-01-01"
end_date = "2022-12-31"

# Subset the dataframe
temp_df = temp_df[(temp_df['DATE'] >= start_date) & (temp_df['DATE'] <= end_date)]
temp_df
```

```python
alt.Chart(temp_df).mark_boxplot().encode(
    alt.X('year(DATE):T', title='Year'), 
    alt.Y('max_min_avg:Q', title='Average Temperature')
)
```
